# 1. 계급 불균형 자료(Imbalanced Data)

계급불균형 자료는 분류 문제에 있어서 Target변수의 분포가 불균형한 경우를 말한다.
분포가 불균형하다는 것은 아래의 그림처럼 타겟변수의 계급 균형비가 동일하지 않고 한쪽에 극심히 치우쳐져 있음을 의미한다.
이러한 문제는 다양한 분야에서 발생하고는 하는데 대표적으로 통신사의 고객이탈, 사기탐지, 기업의 부도 예측, 희귀질환 발병 예측 등이 있다. 이렇게 타겟변수의 분포가 불균형하게 되면 데이터 분석에 있어서 치명적인 문제점이 발생할 우려가 있다.

# 2. 정확도의 역설(Accuracy Paradox)

대부분의 분류 알고리즘들은 정분류율 즉, 얼마나 정확하게 주어진 데이터를 잘 분류하는지에 목적을 두고있다. 이 때 타겟변수의 불균형한 분포가 데이터 분석에 큰 독이 될 수 있다.

어떤 데이터에서 타겟변수의 각 계급의 균형비가 Event : Non-Event = 5 : 95라고 가정해보자. 만약 우리가 생성한 모델이 주어진 모든 경우를 Non-Event라고 예측할 경우, 정분류율이 95%라는 굉장히 높은 수치가 나오게 된다. 굉장히 좋은 성능을 가진 모델처럼 보이지만, 사실 이는 함정을 가지고 있다. 만약 이 모형이 희귀질환의 발병 예측에 사용되었을 경우, 95%라는 높은 정분류율을 자랑하지만 정작 실제로 희귀질환이 발병한 케이스는 단 한 건도 맞추지 못한다. 핵심을 놓치는 모형이 되는 것이다. 이러한 현상을 정확도의 역설(Accuracy Paradox)라고 한다.

계급 불균형 문제를 해결하는 가장 대표적인 방법으로는 샘플링(Sampling) 기법이 있다. 샘플링 기법이란, 데이터셋을 분석에 이용하기 전에 앞서 타겟변수의 계급 비율을 조절해주는 방법을 의미한다. 모형이 두 계급에 대해 비슷한 수준의 학습을 진행할 수 있도록 해주는 것이다.

대표적으로 다수집단의 수를 소수집단만큼 줄이는 방법과 소수집단의 수를 다수집단만큼 늘리는 방법이 있다. 그 외에도 다수집단의 수를 절반만 줄이고 소수집단의 수를 줄어든 다수집단의 수만큼 늘리는 방법도 있다. 정확한 명칭을 살펴보면 Under Sampling, Over Sampling, Under&Over Sampling, SMOTE Sampling 기법이 있다.

# 3. Under Sampling

Under Sampling은 다수 계급에 해당하는 데이터를 랜덤하게 소수 계급의 데이터 수만큼 추출하여 모형을 생성하는 방법이다. 즉, 다수 계급의 수를 줄이는 기법을 의미한다. 이는 비교적 정보의 왜곡이 적게 발생한다는 장점을 가지고 있는데, 없는 데이터를 새로 생성하는 것이 아니라 기존의 데이터만을 가지고 작업이 이루어지기 때문이다. 그러나 조금 더 깊이 생각해보면 Under Sampling 기법은 다수 계급에 해당하는 데이터를 소수 계급의 데이터 수만큼 줄인다는 것이고 이는 곧 그만큼 다수 계급에 대한 정보가 대량 유실된다는 단점을 가지고 있음을 의미한다.

# 4. Over Sampling

Over Sampling 기법은 아래와 같이 크게 두 가지로 나누어진다.
 
1. 단순임의추출에 의한 Over Sampling
2. 복원추출에 의한 Over Sampling
 
먼저 단숨임의추출에 의한 방법은 소수계급에 해당하는 관측치는 모두 이용하고, 다수 계급의 관측치를 소수계급보다 더 적게 임의 추출하는 것이다. Over Sampling이라는 이름을 보고 단순하게 Under Sampling과 반대로 소수계급에 해당하는 관측치를 다수계급만큼 늘리는 기법이라고만 생각할수도 있는데, 단순임의추출에 의한 Over Sampling의 경우 그게 아니라 다수계급의 데이터를 Under Sampling과 유사하게 줄이는 것이다.
그렇기 때문에 오히려 다수계급에 대한 정보 손실은 Under-Sampling 보다도 더 크다. 하지만 소수계급에 대한 가중치가 증가하는 효과를 얻을 수 있다는 장점을 가지고 있다.



다음으로 복원추출에 의한 Over Sampling 기법이 있다. 이 기법은 소수계급에 포함된 샘플을 랜덤하게 다수 계급의 관측치 수만큼 복원 추출하는 것이다. 즉 다수계급의 데이터가 1만개, 소수계급의 데이터가 100개 있다면 소수계급의 데이터를 복원추출해서 1만개까지 늘리는 것이다. 이렇게 할 경우 샘플링 결과, 소수계급 데이터에 중복이 발생하여서 편향이 생기게 된다는 단점을 가지고 있지만, 다수계급의 정보는 손실되지 않는다는 장점을 가지고 있다.



# 5. Under&Over Sampling

Under&Over-Sampling 기법은 Over Sampling  기법과 Under Sampling 기법을 결합한 방법으로 이 방법의 경우 Under Sampling 기법보다 정보의 손실을  절반으로 줄여줄 뿐만 아니라 편향 역시 복원추출에 의한 Over Sampling보다 적다는 특징을 가지고 있다. 전체 표본수를 기준 샘플 수로 설정하고 절반에 해당하는 부분은 소수계급에서 복원추출을 통해 생성한다. 그리고 나머지 절반에 해당하는 부분은 다수계급에서 랜덤추출한다.

예를 들어 아래 그림과 같이 1000개의 샘플이 있는 경우 다수집단(800명)에서 500명을 추출하고, 소수집단(200명)에서 복원추출 방법을 이용해 500명을 추출하게 된다. 이렇게 하면 소수집단과 다수집단간의 비율이 정확하게 1:1가 되기 때문에 다수집단에 편향된 학습을 막을 수 있다.



하지만 이 방법 역시 없는 표본을 일부 새롭게 생성하고, 기존에 있는 표본을 일부 제거하는 것이기 때문에 결과적으로는 정보의 손실과 편향이 동시에 발생한다는 사실은 변하지 않는다. 대신 그 정도가 기존의 Under 또는 Over 샘플링보다는 덜하다는 장점이 있다.

# 6. SMOTE Sampling

SMOTE 기법은 소수계급의 개체수를 늘리는 방법으로 복원추출에 의한 Over Sampling과 유사한 방법이지만 조금은 다른 기법이다.
소수계급의 개체수를 늘리면서 Over Sampling의 치명적인 단점인 데이터의 편향은 막을 수 있는 기법이다.

Over Sampling처럼 똑같은 샘플을 복원추출하여 재생성하는 것이 아니라 기존 샘플들을 조합하여 기존에 없던 새로운 샘플을 만들어낸다. 우선 기준이 되는 샘플을 한 개 잡고 해당 샘플과 가장 가까운 거리에 있는 k개의 샘플을 선택한다. 이때 k개의 샘플은 모두 소수계급에서 추출한다. 그 다음 k개의 샘플 가운데 미리 지정한 샘플링 비율에 따라서 적당한 개수의 샘플을 랜던하게 선별한다. 그리고 각각의 랜덤하게 선별된 샘플과 처음 잡았던 기준 샘플과의 연결 선상의 임의의 점을 선택해 새로운 샘플을 만들어낸다. 글로 표현하면 복잡한데 아래 사진과 함께 보면 이해하기 쉽다.

### Sampling 미진행 시

```python
#Train:Test = 6:4로 데이터 분할
y = credit.Class
x = credit.drop('Class', axis=1)
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.4, random_state = 321, shuffle=True, stratify=y)
```

```python
Dtree = DecisionTreeClassifier()
Dtree.fit(X_train, Y_train)
y_pred = Dtree.predict(X_test)
acc_Dtree = round(accuracy_score(y_pred, Y_test) * 100, 2)
print(acc_Dtree)
```

```python
classificationSummary(Y_test, Dtree.predict(X_test))
print()

#민감도(Sensitivity): 실제로 1인 것 중에 1로 예측한 것의 비율
#특이도(Specificity): 실제로 0인 것 중에 0으로 예측한 것의 비율

cm = confusion_matrix(Y_test, y_pred)

sensitivity = cm[1,1]/(cm[1,0]+cm[1,1])
print('Sensitivity : ', sensitivity )

specificity = cm[0,0]/(cm[0,0]+cm[0,1])
print('Specificity : ', specificity)
```

Sampling을 진행하지 않은 데이터를 6:4로 분할한 후 의사결정나무를 통해 분류 진행 시,

- 정확도: 0.999
- 민감도: 0.741
- 특이도: 0.999

### Under Sampling 진행 시

```python
#다수 계급의 수를 소수 계급의 수만큼 줄이는 Under Sampling 기법
x_us, y_us = RandomUnderSampler(random_state=0).fit_resample(x, y)
```

```python
#Under Sampling 후 0이 284315개 -> 492개로, 1은 492개로 유지되었음을 확인
count0 = 0
count1 = 0
for i in range(len(y_us)):
    if y_us[i] == '0':
        count0 += 1
    else:
        count1 += 1

print('0: ', count0,'개  ', '1: ', count1,'개', sep='')
```

```python
#Train:Test = 6:4로 데이터 분할
X_train_us, X_test_us, Y_train_us, Y_test_us = train_test_split(x_us, y_us, test_size=0.4, random_state = 321, shuffle=True, stratify=y_us)
```

```python
Dtree_us = DecisionTreeClassifier()
Dtree_us.fit(X_train_us, Y_train_us)
y_pred_us = Dtree_us.predict(X_test_us)
acc_Dtree_us = round(accuracy_score(y_pred_us, Y_test_us) * 100, 2)
print(acc_Dtree_us)
```

```python
classificationSummary(Y_test_us, Dtree_us.predict(X_test_us))
print()

#민감도(Sensitivity): 실제로 1인 것 중에 1로 예측한 것의 비율
#특이도(Specificity): 실제로 0인 것 중에 0으로 예측한 것의 비율

cm_us = confusion_matrix(Y_test_us, y_pred_us)

sensitivity_us = cm_us[1,1]/(cm_us[1,0]+cm_us[1,1])
print('Sensitivity : ', sensitivity_us )

specificity_us = cm_us[0,0]/(cm_us[0,0]+cm_us[0,1])
print('Specificity : ', specificity_us)
```

Under Sampling을 진행한 데이터를 6:4로 분할한 후 의사결정나무를 통해 분류 진행 시,

- 정확도: 0.876
- 민감도: 0.883
- 특이도: 0.868

특이도의 경우 Sampling 전보다 떨어지는 경향을 보였지만, 1을 맞추는 민감도는 상승함

### Over Sampling 진행 시 (복원 추출)

```python
#소수 계급의 수를 다수 계급의 수만큼 복원추출 Over Sampling 기법
x_os, y_os = RandomOverSampler(random_state=0).fit_resample(x, y)
```

```python
#Over Sampling 후 0이 284315개로 유지되고, 1은 492 -> 284315개로 복원 추출되었음을 확인
count0 = 0
count1 = 0
for i in range(len(y_os)):
    if y_os[i] == '0':
        count0 += 1
    else:
        count1 += 1

print('0: ', count0,'개  ', '1: ', count1,'개', sep='')
```

```python
#Train:Test = 6:4로 데이터 분할
X_train_os, X_test_os, Y_train_os, Y_test_os = train_test_split(x_os, y_os, test_size=0.4, random_state = 321, shuffle=True, stratify=y_os)
```

```python
Dtree_os = DecisionTreeClassifier()
Dtree_os.fit(X_train_os, Y_train_os)
y_pred_os = Dtree_os.predict(X_test_os)
acc_Dtree_os = round(accuracy_score(y_pred_os, Y_test_os) * 100, 2)
print(acc_Dtree_os)
```

```python
classificationSummary(Y_test_os, Dtree_os.predict(X_test_os))
print()

#민감도(Sensitivity): 실제로 1인 것 중에 1로 예측한 것의 비율
#특이도(Specificity): 실제로 0인 것 중에 0으로 예측한 것의 비율

cm_os = confusion_matrix(Y_test_os, y_pred_os)

sensitivity_os = cm_os[1,1]/(cm_os[1,0]+cm_os[1,1])
print('Sensitivity : ', sensitivity_os )

specificity_os = cm_os[0,0]/(cm_os[0,0]+cm_os[0,1])
print('Specificity : ', specificity_os)
```

Over Sampling을 진행한 데이터를 6:4로 분할한 후 의사결정나무를 통해 분류 진행 시,

- 정확도: 0.999
- 민감도: 1.000
- 특이도: 0.999

1을 정확히 맞추는 민감도는 1.000이, 특이도의 경우에도 0.999의 굉장히 높은 수치가 나타남

### SMOTE Sampling 진행 시

```python
#소수 계급의 수를 다수 계급의 수만큼 새롭게 추가 생성하는 SMOTE Sampling 기법
x_sm, y_sm = SMOTE(random_state=0).fit_resample(x, y)
```

```python
#SMOTE Sampling 후 0이 284315개로 유지되고, 1은 492 -> 284315개로 새롭게 추가 생성되었음을 확인
count0 = 0
count1 = 0
for i in range(len(y_sm)):
    if y_sm[i] == '0':
        count0 += 1
    else:
        count1 += 1

print('0: ', count0,'개  ', '1: ', count1,'개', sep='')
```

```python
#Train:Test = 6:4로 데이터 분할
X_train_sm, X_test_sm, Y_train_sm, Y_test_sm = train_test_split(x_sm, y_sm, test_size=0.4, random_state = 321, shuffle=True, stratify=y_sm)
```

```python
Dtree_sm = DecisionTreeClassifier()
Dtree_sm.fit(X_train_sm, Y_train_sm)
y_pred_sm = Dtree_sm.predict(X_test_sm)
acc_Dtree_sm = round(accuracy_score(y_pred_sm, Y_test_sm) * 100, 2)
print(acc_Dtree_sm)
```

```python
classificationSummary(Y_test_sm, Dtree_sm.predict(X_test_sm))
print()

#민감도(Sensitivity): 실제로 1인 것 중에 1로 예측한 것의 비율
#특이도(Specificity): 실제로 0인 것 중에 0으로 예측한 것의 비율

cm_sm = confusion_matrix(Y_test_sm, y_pred_sm)

sensitivity_sm = cm_sm[1,1]/(cm_sm[1,0]+cm_sm[1,1])
print('Sensitivity : ', sensitivity_sm )

specificity_sm = cm_sm[0,0]/(cm_sm[0,0]+cm_sm[0,1])
print('Specificity : ', specificity_sm)
```
